There are three python types available for the `|` operator:

+ Box2d
+ Text
+ Mask

```python
from ligavision.conf import set_option, CONF_RIKAI_NOTEBOOK_PLATFORM
set_option(CONF_RIKAI_NOTEBOOK_PLATFORM, "github")
```

```python
from ligavision.dsl import Image, Box2d, Mask
from ligavision.viz import Text

uri = 'https://i.scdn.co/image/ab67616d0000b273466def3ce70d94dcacb13c8d'
image = Image(uri)
image * 0.5
```

```python
(image | Box2d(118, 18, 520, 136)).to_image() * 0.5
```

```python
(image | Box2d(118, 18, 520, 136) @ {"color": "#FFF6B0"}).to_image() * 0.5
```

```python
(image | Box2d(118, 18, 520, 136) | Text("Mojito", (118, 8))).to_image() * 0.5
```

```python
image | Box2d(118, 18, 520, 136) | Text("Mojito", (118, 8)) @ {"color": "#000000"}
```

```python
mask = Mask.from_polygon([[118, 18, 118, 136, 520, 136, 520, 18]], image.to_pil().width, image.to_pil().height)
image | mask
```

```python
image | mask @ {"color": "#FFF6B0"}
```
