# Liga Vision DSL for Overlay Attributes
`@` is used for attributes of image overlays.

## Syntax
```python
image | box2d @ {"color": "#000000"}
image | mask @ {"color": "#FFF6B0"}
image | text @ {"color": "#000000"} | box2d 
```

````python
from ligavision.dsl import conf
conf.image.notebook = "github"
````

````python
from ligavision.dsl import Image, Box2d, Mask, Text
uri = 'https://i.scdn.co/image/ab67616d0000b273466def3ce70d94dcacb13c8d'
image = Image(uri)
````

## Change the color of the Text

````python
text_color = image | Box2d(118, 18, 520, 136) | Text("Mojito", (118, 8)) @ {"color": "#000000"}
text_color.to_image() * 0.5
````

## Change the color of the Box2d

````python
box_color = image | Box2d(118, 18, 520, 136) @ {"color": "#00FF00"} | Text("Mojito", (118, 8))
box_color.to_image() * 0.5
````

## Change the color of Mask

````python
mask = Mask.from_polygon([[118, 18, 118, 136, 520, 136, 520, 18]], image.to_pil().width, image.to_pil().height)
masked = image | mask
masked.to_image() * 0.5
````

````python
mask_color = image | mask @ {"color": "#FFF6B0"}
mask_color.to_image() * 0.5
````
