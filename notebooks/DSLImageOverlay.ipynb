# Liga Vision DSL for Image Overlays
`|` is used for image overlays.

## Syntax
```python
image | box2d
image | text
image | mask
image | box2d | text
```

````python
from ligavision.dsl import conf
conf.image.notebook = "github"
````

````python
from ligavision.dsl import Image, Box2d, Mask, Text
uri = 'https://i.scdn.co/image/ab67616d0000b273466def3ce70d94dcacb13c8d'
image = Image(uri)
````

## `|` for image overlays like Box2d and Text

````python
image_1 = image | Box2d(118, 18, 520, 136)
image_1.to_image() * 0.5
````

````python
image_2 = image | Text("Mojito", (118, 8))
image_2.to_image() * 0.5
````

## `|` for multiple image overlays

````python
image_3 = image | Box2d(118, 18, 520, 136) | Text("Mojito", (118, 8))
image_3.to_image() * 0.5
````

## Change the color of the Text

````python
image_3_text_color = image | Box2d(118, 18, 520, 136) | Text("Mojito", (118, 8)) @ {"color": "#000000"}
image_3_text_color.to_image() * 0.5
````

## The Mask image overlay

````python
mask = Mask.from_polygon([[118, 18, 118, 136, 520, 136, 520, 18]], image.to_pil().width, image.to_pil().height)
image_5 = image | mask
image_5.to_image() * 0.5
````

````python
image_6 = image | mask @ {"color": "#FFF6B0"}
image_6.to_image() * 0.5
````
