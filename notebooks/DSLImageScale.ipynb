# Liga Vision DSL for Image Scaling
`*` is used for image scaling.

## Syntax
```python
image * scale_factor
image * (width_scale_factor, height_scale_factor)
```

## Example
``` python
image * 0.5
image * 1.1
image * (0.8, 0.2)
image * (0.2, 0.8)
```
